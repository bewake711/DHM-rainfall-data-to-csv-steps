```python
import pandas as pd
import datetime
import os

os.chdir(r"C:\Users\dell\Desktop\Narayani_Data_Text_Files\Rainfall\0902")  ###### full folder path
start_time = datetime.datetime(1993,1,1)                     #### start time
end_time = datetime.datetime(2005,12,31)                   #### end time
date_range = pd.date_range(start=start_time,end=end_time,freq='D')
print(date_range)

count=1
df_list = []
for file in range(1993,2005+1,1):
    file_name = "AS0902"+str(file)+".txt"
    print(file_name)
    #df_no = "df_"+str(file)
    print(df_no)
    df_no = pd.read_csv(file_name,delim_whitespace=True,header=None)
    del df_no[0]
    df_no.drop(df_no.tail(1).index,inplace=True)
    df_list.append(df_no)
    count=count+1
    #display(df_no)
    
df = pd.concat(df_list,ignore_index = True)
df['Date'] = date_range

display(df)
```

```python
df.to_csv(file_name[:-4]+".csv",index=False,header=False)
cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
df.rename({1: 'Rainfall'}, axis=1, inplace=True)
display(df)
df.to_csv('Station0902.csv',header=False,index=False)    ##################  Name Here   ###################
```
