```python
import pandas as pd

file= '619_Ghorepani.xlsx'
pd.set_option('display.max_rows', 10)

df=pd.read_excel(file,parse_dates=['Date'])
filt = df['Rainfall']=='DNA'
filtered = df[filt].copy()       ############ do this to make a separate copy of dataframe 

# print(filtered._is_view)
# print(filtered._is_copy)
# print(df._is_view)
# print(df._is_copy)

filtered['Year'] = filtered['Date'].dt.year
filtered['Month'] = filtered['Date'].dt.strftime('%m')
filtered['Day'] = filtered['Date'].dt.strftime('%d')
filtered.drop(['gsid','index_no','station','district','year','month','days','Rainfall'], axis=1,inplace=True)
display(filtered)   
```

```python
################## extract missing dates ########
df1=filtered.groupby(by =['Year','Month'],sort=False)
months_name ={1:'Jan',2:'Feb', 3:'Mar',4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
print(months_name[1])

print(df1['Year'])
keys = df1.groups.keys()
for a,b in keys:                         #### a is year, b is month
    #print(a,b)
    #print(a,b, '--->',df1.get_group((a,b))['Day'].to_list())       # string change to inter below
    print(a,months_name[int(b)], '--->',list(map(int, df1.get_group((a,b))['Day'].to_list())))
```

```python
################## extract missing dates in short form ########

def range_maker(input):
    output = []
    current_range, input = input[:1], input[1:]

    def commit_range(output, current_range):
        if len(current_range) < 3:
            # If the range has less than three elements, we just add
            # the individual elements to the output.
            output.extend(current_range)
        else:
            # If the range has three or more elements, we add a tuple
            # to the output containing its lower and upper bound.
            output.append((current_range[0], current_range[-1]))

    for element in input:
        if current_range[-1]+1 == element:
            # The new elements continues the current range, so just add it.
            current_range.append(element)
        else:
            # The new element does not continue the current range, so add
            # the range to the output and start a new one.
            commit_range(output, current_range)
            current_range = [element]

    commit_range(output, current_range)

    return output


df1=filtered.groupby(by =['Year','Month'])
#months_name =['Jan','Feb', 'Mar','Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

print(df1['Year'])
keys = df1.groups.keys()
for a,b in keys:
    #print(a,b)
    #print(a,b, '--->',df1.get_group((a,b))['Day'].to_list())       # string change to inter below
    print(a,months_name[int(b)], '--->',range_maker(list(map(int, df1.get_group((a,b))['Day'].to_list()))))
```
