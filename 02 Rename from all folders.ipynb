```python
import os

os.chdir(r"C:\Users\dell\Desktop\Narayani_Data_Text_Files\Rainfall")
folders_list =[]

for folder in os.listdir():
    if (folder != '.ipynb_checkpoints' and  folder != '02 Rename from all folders.ipynb'):
        folders_list.append(folder)
print('list of folders =', folders_list)
length = len(folders_list)

for num in range(0,length,1):
    folder_path = r"C:\Users\dell\Desktop\Narayani_Data_Text_Files\Rainfall"+"\\"+folders_list[num]
    print(folder_path)
    os.chdir(folder_path)
    for file in os.listdir():
        if (file != '.ipynb_checkpoints' and  file != 'Rename file.ipynb'):
            #print(file)
            temp= file.split('.')
            if len(temp) == 2:
                file_new = temp[0]+temp[1]
            else:
                file_new = temp[0]
            os.rename(file,file_new+'.txt')
    print('done \n')
```
