```python
import os

for file in os.listdir():
    if (file != '.ipynb_checkpoints' and  file != 'Rename file to by adding txt extension.ipynb'):
        print("Old name = ",file)
        temp= file.split('.')
        file_new = temp[0]+temp[1]
        os.rename(file,file_new+'.txt')
        print("New name = ",file_new+'.txt','\n')
```
